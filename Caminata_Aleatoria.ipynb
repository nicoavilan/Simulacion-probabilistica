```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.animation as animation
from IPython.display import HTML
```

```python
N = 10000
x = np.random.rand(N).round(2)
#print(x)
moneda = []
for n in range(N):
  if x[n] > 1/2:
    moneda.append("s")
  else:
    moneda.append("c")
print("sellos: ",moneda.count("s"))
print("caras: ",moneda.count("c"))

p = [0]
for n in range(N):
  if moneda[n] == "c":
    p.append(p[n]+1)
  else:
    p.append(p[n]-1)
plt.figure(figsize= (10,4))
plt.subplot(1,2,1)
sns.countplot(x = moneda)
plt.subplot(1,2,2)
m = np.arange(N+1)
z = np.zeros(N+1)
plt.plot(p, m)
plt.show()
print("Posici√≥n final: ",p[-1])
```

```python
fig=plt.figure()
ax=fig.gca()

def actualizar(i):
  ax.clear()

  ax.plot(p[:i], m[:i],'-')
  ax.plot(p[:i], z[:i],'-')
  ax.set_xlabel("Posici√≥n", fontweight="bold", size = 15)
  ax.set_ylabel("Tiempo", fontweight="bold", size = 15)
  ax.plot(p[i],  m[i],'o',markersize=10,color='r', label = "Evoluci√≥n en\nel tiempo")
  ax.plot(p[i],  0,'o',markersize=10,color='g', label = "Posici√≥n")
  ax.legend()
```

```python
ani=animation.FuncAnimation(fig, actualizar, interval=400)
HTML(ani.to_html5_video())
```
